```python
# Make a summary table
summary_df = pd.DataFrame(purchase_count)
summary_df = summary_df.rename(columns={"Purchase ID": "Purchase Count"})
summary_df['Average Purchase Price'] = avg_purchase_price.map("${:.2f}".format)
summary_df['Total Purchase Value'] = total_purchase_value.map("${:.2f}".format)
summary_df['Avg Total Purchase per Person'] = totals_df_summary['Price'].map("${:.2f}".format)
summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
maxage = heros_data['Age'].max()
agebins = [0,9,14,19,24,29,34,39,maxage]
binlabels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

heros_data['Age Category'] = pd.cut(heros_data['Age'], agebins, labels=binlabels)
player_and_age_grouping = heros_data.groupby(['SN','Age Category'])
player_and_age = pd.DataFrame(player_and_age_grouping['Age Category'].count())
player_and_age = player_and_age.rename(columns={"Age Category": "Total Count"})
player_and_age = player_and_age.reset_index().groupby('Age Category').count()
player_and_age['Percentage of Players'] = (player_and_age['Total Count'] / num_players).map("{:.2%}".format)
player_and_age[["Total Count", "Percentage of Players"]]
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_cat_grouping = heros_data.groupby('Age Category')
purchase_count = age_cat_grouping['Purchase ID'].count()
avg_purchase_price = age_cat_grouping['Price'].mean()
total_purchase_value = age_cat_grouping['Price'].sum()
num_unique_players_per_age_cat = player_and_age_grouping['Purchase ID'].count().reset_index().groupby('Age Category').count()['Purchase ID']
avg_total_purchase_per_person = age_cat_grouping['Price'].sum() / num_unique_players_per_age_cat

list_of_series = [purchase_count, avg_purchase_price, total_purchase_value, avg_total_purchase_per_person]
summary_df = pd.concat(list_of_series, axis=1)
summary_df.columns=[
    'Purchase Count',
        'Average Purchase Price',
            'Total Purchase Value',
            'Avg Total Purchase per Person'
]
summary_df['Average Purchase Price'] = summary_df['Average Purchase Price'].map("${:.2f}".format)
summary_df['Total Purchase Value'] = summary_df['Total Purchase Value'].map("${:.2f}".format)
summary_df['Avg Total Purchase per Person'] = summary_df['Avg Total Purchase per Person'].map("${:.2f}".format)
summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
newdf = heros_data[['Item ID', 'Item Name', 'Price']]
item_group = newdf.groupby(['Item ID','Item Name'])
purchase_count = item_group['Item ID'].count()
total_purchase_value = item_group['Price'].sum()
item_price = total_purchase_value / purchase_count
list_of_series = [purchase_count, item_price, total_purchase_value]
summary_df = pd.concat(list_of_series, axis=1)
summary_df.columns=[
    'Purchase Count',
    'Item Price',
    'Total Purchase Value'
]
summary_df = summary_df.sort_values('Purchase Count', ascending=False)
summary_df['Item Price'] = summary_df['Item Price'].map("${:.2f}".format)
summary_df['Total Purchase Value'] = summary_df['Total Purchase Value'].map("${:.2f}".format)
summary_df.head()
```
